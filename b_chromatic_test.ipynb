```python
from b_chromatic_colouring import *
from b_chromatic_utils import *
from timeit import default_timer as timer
from IPython.display import clear_output
from scipy.sparse.csgraph import minimum_spanning_tree
```

```python
n_vertices_list = [i for i in range(100,1100,100)]+[j for j in range(1500,10500,500)]
```

Testing time when generating MST trees

```python
mst_times = []
for n in n_vertices_list:
    t = timer()
    Ti = random_mst_tree(n)
    t = timer()-t
    mst_times.append(t)
    print("{} -> {}".format(n, t))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1)
ax.set_xlabel("Number of vertices")
ax.set_ylabel("Execution time in seconds")
ax.plot(n_vertices_list, mst_times, "-go", label="Random Tree via MST")
ax.legend()
fig.savefig("random_tree_mst_execution_time.pdf")
```

```python
bfs_times = []
for n in n_vertices_list:
    t = timer()
    Ti = random_bfs_tree(n)
    t = timer()-t
    bfs_times.append(t)
    print("{} -> {}".format(n, t))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1)
ax.set_xlabel("Number of vertices")
ax.set_ylabel("Execution time in seconds")
ax.plot(n_vertices_list, bfs_times, "-go", label="Random Tree via BFS")
ax.legend()
fig.savefig("random_tree_bfs_execution_time.pdf")
```

```python
rtrees_times = []
for n in n_vertices_list:
    t = timer()
    Ti = random_recursive_tree(n)
    t = timer()-t
    rtrees_times.append(t)
    print("{} -> {}".format(n, t))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1)
ax.set_xlabel("Number of vertices")
ax.set_ylabel("Execution time in seconds")
ax.plot(n_vertices_list, rtrees_times, "-go", label="Recursive Random Tree")
ax.legend()
fig.savefig("recursive_random_tree_execution_time.pdf")
```

```python
prufer_trees_times = []
for n in n_vertices_list:
    t = timer()
    Ti = nx.random_tree(n)
    t = timer()-t
    prufer_trees_times.append(t)
    print("{} -> {}".format(n, t))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1)
ax.set_xlabel("Number of vertices")
ax.set_ylabel("Execution time in seconds")
ax.plot(n_vertices_list, prufer_trees_times, "-go", label="Prfer Random Tree")
ax.legend()
fig.savefig("prufer_tree_execution_time.pdf")
```

We'll check on the acumulated time for generating the trees

```python
added_mst_times = [sum(mst_times[:i]) for i in range(len(mst_times))]
added_bfs_times = [sum(bfs_times[:i]) for i in range(len(bfs_times))]
added_rtrees_times = [sum(rtrees_times[:i]) for i in range(len(rtrees_times))]
added_prufer_times = [sum(prufer_trees_times[:i]) for i in range(len(prufer_trees_times))]
```

```python
fig, ax = plt.subplots(1)
ax.set_xlabel("Number of vertices")
ax.set_ylabel("Cumulative Execution time in seconds")
ax.plot(n_vertices_list, added_mst_times, "-ro", label="MST Random Tree")
ax.plot(n_vertices_list, added_bfs_times, "-bo", label="BFS Random Tree")
ax.plot(n_vertices_list, added_rtrees_times, "-yo", label="Recursive Random Tree")
ax.plot(n_vertices_list, added_prufer_times, "-go", label="Prfer Random Tree")
ax.legend()
fig.savefig("cumulative_execution_times.pdf")
```
